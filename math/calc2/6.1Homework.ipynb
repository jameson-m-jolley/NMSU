### Problem 1
Set up an integral for the area of the shaded region. Evaluate the integral to find the area of the shaded region.

**Given:**
- Curve: $y = 5x - x^2$

**Graph Description:**
- The curve enters the window in the third quadrant, goes up and right, changes direction at $(2.5, 6.3)$, and exits in the fourth quadrant.
- The line enters the window in the third quadrant, goes up and right, and exits in the first quadrant.
- The region is below the curve and above the line.

**Task:**
Set up the integral and find the area of the shaded region.

$x=5x-x^2$

$0 = -x^2 - x + 5x $

$0= -1(x^2 + x - 5x)$

$0 = -1(x^2 - 4x)$

$0=-x(x-4)$

range of the integral is [0,4]
we can plot this

```julia
f(x) = x
g(x) = 5*x -x^2

using Plots
r =range(0,4,length=100)

plot(r,f)
plot!(r,g)
```

we then can subtract the 2 and fine the ara by integrating

$(5x-x^2)-(x)$

$\int_{0}^{4} ((5x-x^2)-(x)) \,dx$

$\int_{0}^{4} ((4x-x^2)) \,dx$

```julia
d(x) = (4*x -x^2)
plot!(r,d)
```

$F(x) = 2x^2 - \frac{x^3}{3}$ this is the anti derivative 

so $\int_{0}^{4} ((4x-x^2)) \,dx = (2(4)^2 - \frac{4^3}{3})-2(0)^2 - \frac{(0)^3}{3}$

$ = (2(4)^2 - \frac{4^3}{3})$

$ = (\frac{96}{3}- \frac{64}{3})$

$ = (\frac{32}{3})$
