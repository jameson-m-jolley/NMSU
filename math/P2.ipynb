Evaluate the following limit $\displaystyle \lim_{x \to 0} \ln\left(\left|x\right|\right)\sin\left(x\right)$

```julia
f(x) = log(abs(x))* sin(x)
x= range(-100,100)
plot(x,f,dpi=1000)
```

doing some manipulation we can do this

 $\ln\left(\left|x\right|\right)\sin\left(x\right)$

$=\frac{x\ln\left(\left|x\right|\right)\sin\left(x\right)}{x}$

$=x\ln\left(\left|x\right|\right)\cdot\frac{\sin\left(x\right)}{x}$

$\displaystyle \lim_{x \to 0}\frac{\sin\left(x\right)}{x} = 1$

so by that we have $x\ln\left(\left|x\right|\right)$<br>
we can then get this in the form $\frac{\ln\left(x\right)}{\frac{1}{x}}$

if we do l'hopital's rule we get $\frac{\left(\frac{1}{x}\right)}{\left(-\frac{1}{x^{2}}\right)}$

witch is -x so $\displaystyle \lim_{x \to 0} \ln\left(\left|x\right|\right)\sin\left(x\right) = 0$
